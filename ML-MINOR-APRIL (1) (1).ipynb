```python
#dropping entire column as Product Description consists of all null values
df=df.drop(["Product Description"],axis=1)
```

```python
#checking whether column "Product Description is removed or not"
df.isnull().sum()
```

```python
#replacing Order ZipCode null values with its mean
x=df["Order Zipcode"].mean()
df["Order Zipcode"].replace(np.nan,x,inplace=True)
```

```python
#checking whether NaN values of column " Order Zipcode" are replaced with mean or not
df.isnull().sum()
```

```python
#dropping rows of null values as many of the remaining rows consists of only one null value
df.dropna(inplace=True)
```

```python
#check for null values again 
df.isnull().sum()
```

```python
#check for duplicates 
duplicates=df.duplicated()
print(duplicates.sum())
```

```python
#check for outliers for a particular column(work only for columns having numerical values)using Box Plot analysis 
#Benefit per order has many outliers present
df.boxplot(column=["Benefit per order"])
```

```python
#Removing outliers
#As it consists of  many outliers we removed Benefit per order column
df=df.drop(['Benefit per order'],axis=1)
```

```python
#correlation between numerical columns
df.corr()
```

```python
#bivariate analysis
#pyplot
#for plotting visulaization use matplotlib
import matplotlib.pyplot as plt
x=df["Order Item Quantity"]
y=df["Product Price"]
plt.plot(x,y,marker="H",linewidth=2,markersize=12)
plt.xlabel("Order Item Quantity")
plt.ylabel("Product Price")
plt.show()
```

# Questions

```python
#1.List all the cities and the respective number of pending orders in 'India'?
x=df.loc[(df["Order Country"]=="India")&(df["Order Status"]=="PENDING"),["Order City"]]
print(x.value_counts())
```

```python
#2.Which Country has the most suspected Fraud?
x=df.loc[df["Order Status"]=="SUSPECTED_FRAUD","Order Country"].value_counts()
x.iloc[[0]]
```

```python
#3.Which products(s) has /have the maximum discount?
x=df["Order Item Discount"].max()
x=df.loc[df["Order Item Discount"]==x,["Product Name"]]
x
```

```python
#4.Which year had the minimum sales for "Nike" products?
x=df.loc[(df["Product Name"].str.startswith("Nike")==True),["shipping date (DateOrders)","Sales"]]
z=pd.DataFrame(pd.DatetimeIndex(x["shipping date (DateOrders)"]).year,x["Sales"])
z["shipping date (DateOrders)"].value_counts(ascending=True).index.tolist()[0]
```

```python
#5.How many successfully processed orders have days for shipping(Real)>Days for shipment(Scheduled)?
x=df.loc[((df["Days for shipping (real)"]>df["Days for shipment (scheduled)"])&(df["Order Status"]=="COMPLETE")),["Order Status"]]
x.shape[0]
```
